```python
import matplotlib.pyplot as plt
island_counts = df['island'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(island_counts, labels=island_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Penguins Island')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['body_mass_g'], bins=10)
plt.title('Penguins Body Mass')
plt.show()
```

```python
most_species = df['species'].value_counts().idxmax()
df1 = df[df['species']==most_species]

plt.hist(df1['body_mass_g'], bins=10)
plt.show()
```

```python
df1 = df[df['culmen_length_mm']<=40]
df2 = df1['species'].value_counts()
plt.bar(df2.index, df2.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Penguins Culmen Length')
plt.show()
```
